```python
import folium

# coordenadas das cidades que serão conectadas
coordenadas = [(51.5074, -0.1278), (48.8566, 2.3522), (40.7128, -74.0060)]

# cria o mapa
m = folium.Map(location=[51.5074, -0.1278], zoom_start=4)

# adiciona um marcador para cada cidade
for coordenada in coordenadas:
    folium.Marker(location=coordenada).add_to(m)

# adiciona uma linha que conecta as cidades
folium.PolyLine(locations=coordenadas, color='red').add_to(m)

# exibe o mapa
m
```
